```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import re
```

```python
chrome_options = Options()
chrome_options.add_argument('--headless')
sitemap = []
```

```python
def extract_urls_from_file(file_path):
    with open(file_path, 'r') as file:
        data = file.read()
        # Use regular expression to find all URLs within <loc> tags
        urls = re.findall(r'<loc>(.*?)</loc>', data)
        sitemap.extend(urls)  # Append found URLs to the sitemap list
        return urls

# Call the function with the file path
extracted_urls = extract_urls_from_file('extractprofile.txt')
#print("Extracted URLs:", extracted_urls)
#print("Updated sitemap:", sitemap)
```

```python
def main():
    all_data = []  # Initialize an empty list to store all the data
    
    # Selenium setup
    s = Service("C:/Users/Dell/Desktop/Cricket Final/chromedriver.exe")
    driver = webdriver.Chrome(service=s, options=chrome_options)
    driver.maximize_window()

    for url in extracted_urls:
        driver.get(url)
        #print("\nProcessing:", url)
        
        player_data = {}  # Dictionary to store player's data
        
        try:
            image = driver.find_element(By.XPATH, "//img[@title='profile image']")
            image_source = image.get_attribute('src')
            player_data['Image Source'] = image_source
        except Exception as e:
            player_data['Image Source'] = "Image not found or error occurred"
            
        player_data['Name'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[1]/div[2]/h1""").text
        player_data['Country'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[1]/div[2]/h3""").text
        player_data['Birth Date Header'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[2]""").text
        player_data['Birth Date'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[3]""").text
        player_data['Birth Place Header'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[4]""").text
        player_data['Birth Place'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[5]""").text
        player_data['Height Header'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[6]""").text
        player_data['Height'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[7]""").text
        player_data['Role Header'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[8]""").text
        player_data['Role'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[9]""").text
        
        try:
            player_data['Batting Style Header'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[10]""").text
            player_data['Batting Style'] = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[11]""").text
        except:
            player_data['Batting Style Header'] = "Not Available"
            player_data['Batting Style'] = "Not Available"
            
        try:
            bowling_style_header = driver.find_element(By.XPATH, "/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[12]").text
            if bowling_style_header.strip() == "&nbsp;":
                player_data['Bowling Style Header'] = "Not Available"
            else:
                player_data['Bowling Style Header'] = bowling_style_header
        except:
            player_data['Bowling Style Header'] = "Not Available"
            
        try:
            Bowling_Style = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[13]""").text
            if Bowling_Style.strip() == "&nbsp;":
                player_data['Bowling Style'] = "Not Available"
            else:
                player_data['Bowling Style'] = Bowling_Style
        except:
            player_data['Bowling Style'] = "Not Available"
            
        try:
            Teams_Played_With = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[1]/div/div[30]""").text
            player_data['Teams Played With'] = Teams_Played_With
        except:
            player_data['Teams Played With'] = "Not Available"
        
        try:
            test_debut_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[1]""").text
            player_data['Test Debut Header'] = test_debut_header
        except:
            player_data['Test Debut Header'] = "Not Available"
        
        try:
            Test_Debut = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[2]/a""").text
            player_data['Test Debut'] = Test_Debut
        except:
            player_data['Test Debut'] = "Not Available"
        
        try:
            last_test_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[3]""").text
            player_data['Last Test Header'] = last_test_header
        except:
            player_data['Last Test Header'] = "Not Available"
        
        try:
            Last_Test = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[4]/a""").text
            player_data['Last Test'] = Last_Test
        except:
            player_data['Last Test'] = "Not Available"
        
        try:
            odi_debut_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[5]""").text
            player_data['ODI Debut Header'] = odi_debut_header
        except:
            player_data['ODI Debut Header'] = "Not Available"
        
        try:
            ODI_Debut = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[6]/a""").text
            player_data['ODI Debut'] = ODI_Debut
        except:
            player_data['ODI Debut'] = "Not Available"
        
        try:
            last_odi_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[7]""").text
            player_data['Last ODI Header'] = last_odi_header
        except:
            player_data['Last ODI Header'] = "Not Available"
        
        try:
            Last_ODI = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[8]/a""").text
            player_data['Last ODI'] = Last_ODI
        except:
            player_data['Last ODI'] = "Not Available"
        
        try:
            t20_debut_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[9]""").text
            player_data['T20 Debut Header'] = t20_debut_header
        except:
            player_data['T20 Debut Header'] = "Not Available"
            
        try:
            T20_Debut = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[10]/a""").text
            player_data['T20 Debut'] = T20_Debut
        except:
            player_data['T20 Debut'] = "Not Available"
            
        try:
            last_t20_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[11]""").text
            player_data['Last T20 Header'] = last_t20_header
        except:
            player_data['Last T20 Header'] = "Not Available"
            
        try:
            Last_T20 = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[12]/a""").text
            player_data['Last T20'] = Last_T20
        except:
            player_data['Last T20'] = "Not Available"
            
        try:
            ipl_debut_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[13]""").text
            player_data['IPL Debut Header'] = ipl_debut_header
        except:
            player_data['IPL Debut Header'] = "Not Available"
            
        try:
            IPL_Debut = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[14]/a""").text
            player_data['IPL Debut'] = IPL_Debut
        except:
            player_data['IPL Debut'] = "Not Available"
            
        try:
            last_ipl_header = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[15]""").text
            player_data['Last IPL Header'] = last_ipl_header
        except:
            player_data['Last IPL Header'] = "Not Available"
            
        try:
            Last_IPL = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[5]/div[16]/a""").text
            player_data['Last IPL'] = Last_IPL
        except:
            player_data['Last IPL'] = "Not Available"
            
        try:
            Profile = driver.find_element(By.XPATH,"""/html/body/div[1]/div[2]/div[3]/div[2]/div[2]/div/div[7]""").text
            player_data['Profile'] = Profile
        except:
            player_data['Profile'] = "Not Available"
        
        all_data.append(player_data)  # Append player's data dictionary to the list
        
        time.sleep(3)
    
    driver.quit()
    
    # Print the collected data for all players
    for idx, player in enumerate(all_data[:20],start=1):
        print(f"\nPlayer {idx} Data:")
        for key, value in player.items():
            print(f"{key}: {value}")
    
    return all_data

if __name__ == "__main__":
    main()
```
