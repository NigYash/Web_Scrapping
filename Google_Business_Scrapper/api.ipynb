```python
import googlemaps
import pandas as pd
import time

# Initialize DataFrame
df = pd.DataFrame(columns=['Place Id', 'Name', 'Address', 'Contact', 'URL', 'Lat', 'Lng', 'Service/Business', 'District'])

# List of businesses and districts
businesses = ['parlours', 'salon', 'beauty clinics', 'health care centres', 'gym', 'physical fitness centres', 'massage centre', 'pedicure centre', 'wedding hall', 'restaurants', 'hotel', 'clinics', 'catering', 'shamiana', 'lawns', 'guest houses']
districts = ['Agra', 'Barabanki', 'Lucknow', 'Deoria', 'Basti', 'Barabanki', 'Mathura', 'Meerut']

# API Key
YOUR_API_KEY = 'AIzaSyAwsPpz2WW-xzI6KBjgiCXyCktRr0A34iU'
gmaps = googlemaps.Client(key=YOUR_API_KEY)

# Function to fetch details of places and add to DataFrame
def fetch_place_details(place, business_type, district):
    global counter
    try:
        place_id = place['place_id']
        details = gmaps.place(place_id=place_id)['result']
        contact = details.get('formatted_phone_number', '')
        url = details.get('url', '')
        df.loc[counter] = [place_id, place['name'], place['formatted_address'], contact, url, place['geometry']['location']['lat'], place['geometry']['location']['lng'], business_type, district]
        counter += 1
    except Exception as e:
        print(f"Error fetching details for place {place['name']}: {e}")

# Initialize counter
counter = 0

# Loop through each district and business type
for district in districts:
    for business in businesses:
        query = f"{business} in {district}"
        print(query)
        
        next_page_token = None
        for r in range(4):  # Loop to handle pagination
            try:
                if r == 0:
                    result = gmaps.places(query=query)
                else:
                    if next_page_token:
                        time.sleep(2)  # Wait for the token to become active
                        result = gmaps.places(query=query, page_token=next_page_token)
                    else:
                        break
                
                for place in result['results']:
                    fetch_place_details(place, business, district)
                
                next_page_token = result.get('next_page_token', None)
                if not next_page_token:
                    break
            except Exception as e:
                print(f"Error during query search: {e}")
                break

# Save DataFrame to a CSV file
df.to_csv('business_data.csv', index=False)
print("Data collection complete.")
```
